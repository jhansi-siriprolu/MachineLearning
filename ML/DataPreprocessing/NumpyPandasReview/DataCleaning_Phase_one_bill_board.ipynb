```python
#check if combination of date_e and time is unique for each column
bill_df1 = bill_df.set_index(['date_e','time'])
# len(bill_df['date_e','time'].unique()) == bill_df.shape[0]
```

```python
mlt_df = bill_df1.reset_index().melt(id_vars = ['date_e', 'time', 'artist', 'track'],
              value_vars=bill_df1.columns[2:],
              var_name="Week_num",
              value_name="Top100")
```

```python
# In mlt_df, show all of the weeks when the song "Oops!.. I Did It Again" was in the top 100.
bm1 = mlt_df["track"] == 'Oops!.. I Did It Again'
bm2 = mlt_df["Top100"] < 100
bm = (bm1) & (bm2)
mlt_df[bm].Week_num
```
