```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

# Binomial distribution
### Used when only 2 outcomes are possible
### probaility of x success in n trails

### FTO Ecommerce website has 10% return rate.
### What is the probability that exactly 5 customers out of 20 return the item

```python
#param1 : expected number of successful trails
#param2 : total number of trails
#param3 : probability of success
stats.binom.pmf(5,20,0.1)
```

### Probaility that max 5 customers will return

```python
stats.binom.cdf(5,20,0.1)
```

### Probability that more 5 will return

```python
1 - stats.binom.cdf(5,20,0.1)
```

## Avg no of customers who will return, variance and std

```python
Average_no_of_customers_who_return, variance_of_return = stats.binom.stats(20,0.1)
```

```python
print(f"Average_no_of_customers_who_return : {Average_no_of_customers_who_return} \nVariance_of_return: {variance_of_return}")
```

```python
binom_df = pd.DataFrame({'Success':range(1,21),'pmf':list(stats.binom.pmf(range(1,21),20,0.1))})
```

```python
sn.barplot(x = "Success", y = 'pmf', data = binom_df)
plt.title("BINOMIAL PLOT")
```

# Poission

## Number of events that can occur over a period of time

### no of calls at call center follows poisson distribution with 10 calls/hour

```python
print("Probability that Number of calls at call center not greater than 5")
stats.poisson.cdf(5,10)
```

## number of calls will exceed 30 in 3 hour period.

```python
1 - stats.poisson.cdf(30,30)
```

```python
df_poisson = pd.DataFrame({'Success':range(0,30),'pmf': list(stats.poisson.pmf(range(0,30),10))})
```

```python
sn.barplot(x = "Success",y = "pmf", data = df_poisson)
plt.title("Poisson Distribution of call center receiving an average 10 calls in a hour")
```

# Exponential distribution

### Single parameter distribution used in calculating the time to failure of events occurening continously and independently

```python
# Mean time between failures is 1000
# Probability that system will fail before 1000 hours
stats.expon.cdf(1000, scale = 1000,loc = 1/1000)
```

```python
# pro system will not fail until 2000 hours
1 - stats.expon.cdf(2000, loc = 1/1000, scale = 1000)
```

### Percentange point function

```python
# Time by which 10 percent of the system will fail
stats.expon.ppf(0.1,scale = 1000, loc = 1/1000)
```

```python
df_exp = pd.DataFrame({"Sucess": range(0,5000,100),"pdf":list(stats.expon.pdf(range(0,5000,100),loc = 1/1000, scale = 1000))})
```

```python
sn.barplot(x = "Sucess",y = "pdf",data = df_exp)
```
