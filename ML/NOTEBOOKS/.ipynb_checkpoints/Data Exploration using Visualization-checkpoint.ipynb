```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

pd.set_option('display.max_columns',7)
%matplotlib inline
```

```python
df = pd.read_csv("../data/IPL IMB381IPL2013.csv")
```

```python
df_age = df.groupby('AGE')["SOLD PRICE"].mean().reset_index()
```

## BAR PLOT
### Frequency distribution of categorical variable

```python
sn.barplot(x ='AGE',\
           y = 'SOLD PRICE',\
           data = df_age)
```

## STACKED BAR PLOT

```python
df_age_role = df.groupby(['PLAYING ROLE','AGE'])["SOLD PRICE"].mean().reset_index()
```

```python
sn.barplot(x = 'AGE',\
          y = "SOLD PRICE",\
          hue = 'PLAYING ROLE',\
          data = df_age_role)
```

## Histogram
### frequency distribution of continous variable

```python
plt.hist(df["SOLD PRICE"])
#default 10 bins
```

```python
plt.hist(df['SOLD PRICE'],bins = 20)
```

## Density distribution
### Distrinution of data over a continous interval

```python
sn.distplot(df['SOLD PRICE'])
```

## Box/ Whisker plot - variablity of data and existence of outliers

```python
bp = plt.boxplot(df["SOLD PRICE"])
```

```python
min , max = [item.get_ydata()[0] for item in bp['caps']]
```

```python
quartile_25,quartile_75 = [item.get_ydata()[0] for item in bp["whiskers"]]
```

```python
print(f"Minimum value :  {min} \n 25th quartile:  {quartile_25} \n Maximum value:  {max} \n 75th quartile:  {quartile_75}")
```

# FINDING OUTLIERS

```python
df[df['SOLD PRICE'] > max][["PLAYER NAME","PLAYING ROLE","SOLD PRICE"]]
```

## Comparing Distributions

```python
df["sp_captain"] = df[df["CAPTAINCY EXP"] == 1]["SOLD PRICE"]
df["sp_non_captain"] = df[df["CAPTAINCY EXP"] == 0]["SOLD PRICE"]
```

```python
sn.distplot(df.sp_captain,label = "Captain",color = 'r')
sn.distplot(df.sp_non_captain,label = "NON - Captain",color = 'y')
plt.legend()
```

```python
sn.boxplot(x = "PLAYING ROLE", y = "SOLD PRICE",data = df)
```

The SOLD PRICE of Allrounder and Batsman is higher than Bowler and W.keeper. 
The Allrounders who have greater than 1.35L SOLD PRICE are not outlier's but because of the variance in the data.
The outliers in Batsman and W.Keeper can refer to OUTLIERS dataframe above

```python
df_b = df[df["PLAYING ROLE"] == "Batsman"]
```

```python
plt.scatter(x = "SIXERS",\
           y = "SOLD PRICE",\
           data = df_b)
```

```python
sn.regplot(x = "SIXERS",\
           y = "SOLD PRICE",\
           data = df_b)
```
