# MATPLOT LIB

```python
data = pd.read_csv("adult.csv")
data.head()
```

```python
cols = ["age","education-num","capitalGain"]
data[cols]
```

```python

for i, col_name in zip(range(1,4),cols):
    # plt.figure(figsize = (10,4))
    print(i)
    plt.subplot(2,3,i)
    plt.hist(data[col_name],bins = 30)
    plt.title(col_name)
    # plt.show()
    i += 3
    print(i)
    plt.subplot(2,3,i)
    plt.boxplot(data[col_name])
    plt.title(col_name)
    plt.tight_layout()
plt.show()
```
